# Vector Manipulation in Python

In this lab, you'll practice advanced NumPy operations with arrays and matrices. We'll explore scaling, translation, and especially rotation transformations, which are fundamental for understanding how to map between vector spaces (like English to French embeddings).

## 1. Imports
We'll use NumPy for array manipulation and Matplotlib for visualization.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Helper Function: Plotting Vectors
We'll define a helper function to plot vectors as arrows in 2D.

```python
def plot_vectors(vectors, axes=[4,4], colors=['black', 'blue', 'red', 'green'], fname=None):
    plt.figure(figsize=(6,6))
    plt.axvline(x=0, color='grey', lw=1)
    plt.axhline(y=0, color='grey', lw=1)
    for i, vec in enumerate(vectors):
        color = colors[i % len(colors)]
        plt.arrow(0, 0, vec[0,0], vec[0,1], head_width=0.15, head_length=0.2, fc=color, ec=color, length_includes_head=True)
    plt.xlim(-axes[0], axes[0])
    plt.ylim(-axes[1], axes[1])
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    if fname:
        plt.savefig(fname)
    plt.show()
```

## 3. Example 1: Scaling and Rotation
Let's create a 2x2 matrix and a row vector, and see how matrix multiplication transforms the vector.

```python
# Create a 2x2 matrix
R = np.array([[-2, 0],
              [0, 2]])
print('Matrix R:')
print(R)

# Create a row vector
x = np.array([[1, 1]])
print('Vector x:')
print(x)

# Apply the transformation
y = np.dot(R, x.T)
print('Transformed vector y:')
print(y)
```

```python
# Plot x
plot_vectors([x], axes=[4,4])
# Plot x and y
plot_vectors([x, y.T], axes=[4,4])
```

## 4. Example 2: Rotation Matrix in 2D
Let's define a rotation matrix that rotates vectors counterclockwise by 100 degrees.

```python
angle = 100 * (np.pi / 180)  # Convert degrees to radians
Ro = np.array([[np.cos(angle), -np.sin(angle)],
               [np.sin(angle),  np.cos(angle)]])
print('Rotation matrix Ro:')
print(Ro)

x2 = np.array([[2, 2]])
y2 = np.dot(Ro, x2.T)
print('Rotated vector y2:')
print(y2)
print('x2 norm', np.linalg.norm(x2))
print('y2 norm', np.linalg.norm(y2))
print('Rotation matrix norm', np.linalg.norm(Ro))
```

```python
# Plot x2 and y2
plot_vectors([x2, y2.T], axes=[4,4])
```

## 5. Frobenius Norm
The Frobenius norm is a generalization of the vector norm to matrices. Let's compute it for a sample matrix and compare with np.linalg.norm.

```python
A = np.array([[2, 2],
              [2, 2]])
A_squared = np.square(A)
print('A squared:')
print(A_squared)
A_Frobenius = np.sqrt(np.sum(A_squared))
print('Frobenius norm (manual):', A_Frobenius)
print('Frobenius norm (np.linalg.norm):', np.linalg.norm(A))
```

```python
# Frobenius norm of the rotation matrix
print('Frobenius norm of the Rotation matrix')
print(np.sqrt(np.sum(Ro * Ro)), '==', np.linalg.norm(Ro))
```
